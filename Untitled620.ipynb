```python
import os
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor

# ==============================
# PATHS
# ==============================
data_path = r"C:\Users\NXTWAVE\Downloads\Company Failure Risk Prediction\Non-Government-companies-struck-off-State-wise-distribution.csv"
output_dir = r"C:\Users\NXTWAVE\Downloads\Company Failure Risk Prediction"

os.makedirs(output_dir, exist_ok=True)

# ==============================
# LOAD DATA
# ==============================
df = pd.read_csv(data_path)
df = df.dropna()

print("\n✅ Dataset Loaded\n")

# ==============================
# PREPROCESSING
# ==============================
X = df.iloc[:, :-1]
y = df.iloc[:, -1]

# Encode categorical
for col in X.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col])

# Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# ==============================
# BAT ALGORITHM
# ==============================
n_bats = 10
iterations = 10
fmin, fmax = 0, 2
alpha = 0.9
gamma = 0.9

# Initialize bats
positions = np.random.uniform(low=[50, 3], high=[300, 20], size=(n_bats, 2))
velocities = np.zeros((n_bats, 2))
frequencies = np.zeros(n_bats)
loudness = np.ones(n_bats)
pulse_rate = np.random.rand(n_bats)

def fitness(pos):
    model = RandomForestRegressor(
        n_estimators=int(pos[0]),
        max_depth=int(pos[1]),
        random_state=42
    )
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    return r2_score(y_test, preds)

print("\n🦇 Running Bat Algorithm Optimization...\n")

# Initial best
fitness_vals = np.array([fitness(p) for p in positions])
best_index = np.argmax(fitness_vals)
best = positions[best_index].copy()

# ==============================
# BA LOOP
# ==============================
for t in range(iterations):
    for i in range(n_bats):
        frequencies[i] = fmin + (fmax - fmin) * random.random()

        velocities[i] += (positions[i] - best) * frequencies[i]
        new_pos = positions[i] + velocities[i]

        # Local search
        if random.random() > pulse_rate[i]:
            new_pos = best + 0.01 * np.random.randn(2)

        # Bounds
        new_pos[0] = np.clip(new_pos[0], 10, 300)
        new_pos[1] = np.clip(new_pos[1], 2, 20)

        new_fit = fitness(new_pos)

        # Accept new solution
        if new_fit > fitness_vals[i] and random.random() < loudness[i]:
            positions[i] = new_pos
            fitness_vals[i] = new_fit
            loudness[i] *= alpha
            pulse_rate[i] = pulse_rate[i] * (1 - np.exp(-gamma * t))

    # Update global best
    best_index = np.argmax(fitness_vals)
    best = positions[best_index].copy()

    print(f"Iteration {t+1} Best R2: {fitness_vals[best_index]:.4f}")

best_params = best.astype(int)

print("\n🏆 Best Parameters:", best_params)

# ==============================
# FINAL MODEL
# ==============================
model = RandomForestRegressor(
    n_estimators=int(best_params[0]),
    max_depth=int(best_params[1]),
    random_state=42
)

model.fit(X_train, y_train)
y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
accuracy = r2 * 100

# ==============================
# SAVE RESULTS CSV
# ==============================
results_df = pd.DataFrame([{
    "Model": "BA_RandomForest",
    "MSE": mse,
    "R2": r2,
    "Accuracy (%)": accuracy
}])

results_df.to_csv(os.path.join(output_dir, "ba_results.csv"), index=False)

# ==============================
# SAVE PREDICTIONS CSV
# ==============================
pred_df = pd.DataFrame({
    "Actual": y_test.values,
    "Predicted": y_pred
})

pred_df.to_csv(os.path.join(output_dir, "ba_predictions.csv"), index=False)

# ==============================
# SAVE + SHOW FUNCTION
# ==============================
def save_and_show(fig, filename):
    path = os.path.join(output_dir, filename)
    fig.savefig(path)
    plt.show()
    print(f"Saved: {filename}")

# ==============================
# GRAPH 1: ACCURACY
# ==============================
fig1 = plt.figure()
plt.bar(["BA_RF"], [accuracy])
plt.title("Bat Algorithm Optimized Accuracy")
plt.ylabel("Accuracy (%)")

save_and_show(fig1, "ba_accuracy_graph.png")

# ==============================
# GRAPH 2: PREDICTION
# ==============================
fig2 = plt.figure()
plt.plot(pred_df["Actual"].values[:50], label="Actual")
plt.plot(pred_df["Predicted"].values[:50], label="Predicted")

plt.legend()
plt.title("BA Prediction Graph")

save_and_show(fig2, "ba_prediction_graph.png")

# ==============================
# GRAPH 3: HEATMAP
# ==============================
corr = df.corr(numeric_only=True)

fig3 = plt.figure()
plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("BA Correlation Heatmap")

save_and_show(fig3, "ba_heatmap.png")

# ==============================
# GRAPH 4: METRICS
# ==============================
fig4 = plt.figure()
plt.bar(["MSE", "R2"], [mse, r2])

plt.title("BA Model Metrics")

save_and_show(fig4, "ba_metrics_graph.png")

# ==============================
# SAVE JSON SUMMARY
# ==============================
summary = {
    "best_parameters": {
        "n_estimators": int(best_params[0]),
        "max_depth": int(best_params[1])
    },
    "MSE": float(mse),
    "R2": float(r2),
    "Accuracy": float(accuracy)
}

with open(os.path.join(output_dir, "ba_summary.json"), "w") as f:
    json.dump(summary, f, indent=4)

print("\n✅ BA RESULTS GENERATED SUCCESSFULLY!")
```
