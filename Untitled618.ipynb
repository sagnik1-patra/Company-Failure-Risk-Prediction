```python
import os
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor

# ==============================
# PATHS
# ==============================
data_path = r"C:\Users\NXTWAVE\Downloads\Company Failure Risk Prediction\Non-Government-companies-struck-off-State-wise-distribution.csv"
output_dir = r"C:\Users\NXTWAVE\Downloads\Company Failure Risk Prediction"

os.makedirs(output_dir, exist_ok=True)

# ==============================
# LOAD DATA
# ==============================
df = pd.read_csv(data_path)
df = df.dropna()

print("\n✅ Dataset Loaded\n")

# ==============================
# PREPROCESSING
# ==============================
X = df.iloc[:, :-1]
y = df.iloc[:, -1]

# Encode categorical
for col in X.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col])

# Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# ==============================
# CSA IMPLEMENTATION
# ==============================
def create_nest():
    return [random.randint(50, 300), random.randint(3, 20)]

def levy_flight():
    return random.uniform(-1, 1)

def fitness(solution):
    model = RandomForestRegressor(
        n_estimators=int(solution[0]),
        max_depth=int(solution[1]),
        random_state=42
    )
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    return r2_score(y_test, preds)

# ==============================
# CSA PARAMETERS
# ==============================
n_nests = 10
iterations = 10
pa = 0.25  # probability of abandoning nests

nests = [create_nest() for _ in range(n_nests)]

print("\n🐦 Running CSA Optimization...\n")

# ==============================
# CSA LOOP
# ==============================
for i in range(iterations):
    fitness_scores = [fitness(n) for n in nests]

    best_index = np.argmax(fitness_scores)
    best_nest = nests[best_index]

    for j in range(n_nests):
        if random.random() > pa:
            step = levy_flight()
            new_nest = [
                nests[j][0] + int(step * random.randint(-20, 20)),
                nests[j][1] + int(step * random.randint(-3, 3))
            ]

            # bounds
            new_nest[0] = max(10, min(300, new_nest[0]))
            new_nest[1] = max(2, min(20, new_nest[1]))

            if fitness(new_nest) > fitness_scores[j]:
                nests[j] = new_nest

    print(f"Iteration {i+1} Best R2: {fitness(best_nest):.4f}")

best_solution = max(nests, key=lambda x: fitness(x))

print("\n🏆 Best Parameters:", best_solution)

# ==============================
# FINAL MODEL
# ==============================
model = RandomForestRegressor(
    n_estimators=int(best_solution[0]),
    max_depth=int(best_solution[1]),
    random_state=42
)

model.fit(X_train, y_train)
y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
accuracy = r2 * 100

# ==============================
# SAVE RESULTS CSV
# ==============================
results_df = pd.DataFrame([{
    "Model": "CSA_RandomForest",
    "MSE": mse,
    "R2": r2,
    "Accuracy (%)": accuracy
}])

results_df.to_csv(os.path.join(output_dir, "csa_results.csv"), index=False)

# ==============================
# SAVE PREDICTIONS CSV
# ==============================
pred_df = pd.DataFrame({
    "Actual": y_test.values,
    "Predicted": y_pred
})

pred_df.to_csv(os.path.join(output_dir, "csa_predictions.csv"), index=False)

# ==============================
# SAVE + SHOW FUNCTION
# ==============================
def save_and_show(fig, filename):
    path = os.path.join(output_dir, filename)
    fig.savefig(path)
    plt.show()
    print(f"Saved: {filename}")

# ==============================
# GRAPH 1: ACCURACY
# ==============================
fig1 = plt.figure()
plt.bar(["CSA_RF"], [accuracy])
plt.title("CSA Optimized Model Accuracy")
plt.ylabel("Accuracy (%)")

save_and_show(fig1, "csa_accuracy_graph.png")

# ==============================
# GRAPH 2: PREDICTION
# ==============================
fig2 = plt.figure()
plt.plot(pred_df["Actual"].values[:50], label="Actual")
plt.plot(pred_df["Predicted"].values[:50], label="Predicted")

plt.legend()
plt.title("CSA Prediction Graph")

save_and_show(fig2, "csa_prediction_graph.png")

# ==============================
# GRAPH 3: HEATMAP
# ==============================
corr = df.corr(numeric_only=True)

fig3 = plt.figure()
plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("CSA Correlation Heatmap")

save_and_show(fig3, "csa_heatmap.png")

# ==============================
# GRAPH 4: METRICS
# ==============================
fig4 = plt.figure()
plt.bar(["MSE", "R2"], [mse, r2])

plt.title("CSA Model Metrics")

save_and_show(fig4, "csa_metrics_graph.png")

# ==============================
# SAVE JSON
# ==============================
summary = {
    "best_parameters": {
        "n_estimators": int(best_solution[0]),
        "max_depth": int(best_solution[1])
    },
    "MSE": float(mse),
    "R2": float(r2),
    "Accuracy": float(accuracy)
}

with open(os.path.join(output_dir, "csa_summary.json"), "w") as f:
    json.dump(summary, f, indent=4)

print("\n✅ CSA RESULTS GENERATED SUCCESSFULLY!")
```
