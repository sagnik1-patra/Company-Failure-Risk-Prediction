```python
import os
import json
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor

# ==============================
# PATHS
# ==============================
data_path = r"C:\Users\NXTWAVE\Downloads\Company Failure Risk Prediction\Non-Government-companies-struck-off-State-wise-distribution.csv"
output_dir = r"C:\Users\NXTWAVE\Downloads\Company Failure Risk Prediction"

os.makedirs(output_dir, exist_ok=True)

# ==============================
# LOAD DATA
# ==============================
df = pd.read_csv(data_path)
df = df.dropna()

print("\n✅ Dataset Loaded\n")

# ==============================
# PREPROCESSING
# ==============================
X = df.iloc[:, :-1]
y = df.iloc[:, -1]

for col in X.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col])

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# ==============================
# AIS (CLONAL SELECTION)
# ==============================
def create_antibody():
    return {
        "n_estimators": random.randint(50, 300),
        "max_depth": random.randint(3, 20)
    }

def mutate(antibody):
    return {
        "n_estimators": max(10, antibody["n_estimators"] + random.randint(-20, 20)),
        "max_depth": max(2, antibody["max_depth"] + random.randint(-3, 3))
    }

def fitness(antibody):
    model = RandomForestRegressor(
        n_estimators=antibody["n_estimators"],
        max_depth=antibody["max_depth"],
        random_state=42
    )
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    return r2_score(y_test, preds)

# ==============================
# AIS OPTIMIZATION LOOP
# ==============================
population_size = 10
generations = 10
clone_factor = 3

population = [create_antibody() for _ in range(population_size)]

print("\n🧬 Running AIS Optimization...\n")

for gen in range(generations):
    scored = [(ab, fitness(ab)) for ab in population]
    scored.sort(key=lambda x: x[1], reverse=True)

    print(f"Generation {gen+1} Best R2: {scored[0][1]:.4f}")

    # Select top antibodies
    top = scored[:5]

    # Clone + mutate
    new_population = []
    for ab, score in top:
        for _ in range(clone_factor):
            new_population.append(mutate(ab))

    population = new_population

# Best solution
best_antibody = sorted([(ab, fitness(ab)) for ab in population],
                       key=lambda x: x[1], reverse=True)[0][0]

print("\n🏆 Best Parameters:", best_antibody)

# ==============================
# FINAL MODEL
# ==============================
model = RandomForestRegressor(
    n_estimators=best_antibody["n_estimators"],
    max_depth=best_antibody["max_depth"],
    random_state=42
)

model.fit(X_train, y_train)
y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
accuracy = r2 * 100

# ==============================
# RESULTS CSV
# ==============================
results_df = pd.DataFrame([{
    "Model": "AIS_RandomForest",
    "MSE": mse,
    "R2": r2,
    "Accuracy (%)": accuracy
}])

results_df.to_csv(os.path.join(output_dir, "ais_results.csv"), index=False)

# ==============================
# PREDICTIONS CSV
# ==============================
pred_df = pd.DataFrame({
    "Actual": y_test.values,
    "Predicted": y_pred
})

pred_df.to_csv(os.path.join(output_dir, "ais_predictions.csv"), index=False)

# ==============================
# SAVE + SHOW FUNCTION
# ==============================
def save_and_show(fig, filename):
    path = os.path.join(output_dir, filename)
    fig.savefig(path)
    plt.show()
    print(f"Saved: {filename}")

# ==============================
# GRAPH 1: ACCURACY
# ==============================
fig1 = plt.figure()
plt.bar(["AIS_RF"], [accuracy])
plt.title("AIS Optimized Model Accuracy")
plt.ylabel("Accuracy (%)")

save_and_show(fig1, "ais_accuracy_graph.png")

# ==============================
# GRAPH 2: PREDICTION GRAPH
# ==============================
fig2 = plt.figure()
plt.plot(pred_df["Actual"].values[:50], label="Actual")
plt.plot(pred_df["Predicted"].values[:50], label="Predicted")

plt.legend()
plt.title("AIS Prediction Graph")

save_and_show(fig2, "ais_prediction_graph.png")

# ==============================
# GRAPH 3: HEATMAP
# ==============================
corr = df.corr(numeric_only=True)

fig3 = plt.figure()
plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("AIS Correlation Heatmap")

save_and_show(fig3, "ais_heatmap.png")

# ==============================
# GRAPH 4: RESULT BAR
# ==============================
fig4 = plt.figure()
plt.bar(["MSE", "R2"], [mse, r2])

plt.title("AIS Model Metrics")

save_and_show(fig4, "ais_metrics_graph.png")

# ==============================
# SAVE JSON
# ==============================
summary = {
    "best_parameters": best_antibody,
    "MSE": float(mse),
    "R2": float(r2),
    "Accuracy": float(accuracy)
}

with open(os.path.join(output_dir, "ais_summary.json"), "w") as f:
    json.dump(summary, f, indent=4)

print("\n✅ AIS RESULTS GENERATED SUCCESSFULLY!")
```
