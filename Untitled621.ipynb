```python
import os
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor

# ==============================
# PATHS
# ==============================
data_path = r"C:\Users\NXTWAVE\Downloads\Company Failure Risk Prediction\Non-Government-companies-struck-off-State-wise-distribution.csv"
output_dir = r"C:\Users\NXTWAVE\Downloads\Company Failure Risk Prediction"

os.makedirs(output_dir, exist_ok=True)

# ==============================
# LOAD DATA
# ==============================
df = pd.read_csv(data_path)
df = df.dropna()

print("\n✅ Dataset Loaded\n")

# ==============================
# PREPROCESSING
# ==============================
X = df.iloc[:, :-1]
y = df.iloc[:, -1]

# Encode categorical
for col in X.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col])

# Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# ==============================
# ALOA IMPLEMENTATION
# ==============================
def fitness(solution):
    model = RandomForestRegressor(
        n_estimators=int(solution[0]),
        max_depth=int(solution[1]),
        random_state=42
    )
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    return r2_score(y_test, preds)

# Initialize ants and antlions
n_agents = 10
iterations = 10

ants = np.random.uniform([50, 3], [300, 20], (n_agents, 2))
antlions = ants.copy()

print("\n🦁 Running ALOA Optimization...\n")

# ==============================
# ALOA LOOP
# ==============================
for t in range(iterations):
    fitness_vals = np.array([fitness(a) for a in ants])
    antlion_fitness = np.array([fitness(a) for a in antlions])

    best_idx = np.argmax(antlion_fitness)
    elite = antlions[best_idx]

    for i in range(n_agents):
        # random walk influenced by elite
        walk = np.random.randn(2)
        ants[i] = ants[i] + walk * (elite - ants[i])

        # bounds
        ants[i][0] = np.clip(ants[i][0], 10, 300)
        ants[i][1] = np.clip(ants[i][1], 2, 20)

        if fitness(ants[i]) > antlion_fitness[i]:
            antlions[i] = ants[i]

    print(f"Iteration {t+1} Best R2: {np.max(antlion_fitness):.4f}")

best_solution = antlions[np.argmax([fitness(a) for a in antlions])]

print("\n🏆 Best Parameters:", best_solution.astype(int))

# ==============================
# FINAL MODEL
# ==============================
model = RandomForestRegressor(
    n_estimators=int(best_solution[0]),
    max_depth=int(best_solution[1]),
    random_state=42
)

model.fit(X_train, y_train)
y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
accuracy = r2 * 100

# ==============================
# SAVE RESULTS CSV
# ==============================
results_df = pd.DataFrame([{
    "Model": "ALOA_RandomForest",
    "MSE": mse,
    "R2": r2,
    "Accuracy (%)": accuracy
}])

results_df.to_csv(os.path.join(output_dir, "aloa_results.csv"), index=False)

# ==============================
# SAVE PREDICTIONS CSV
# ==============================
pred_df = pd.DataFrame({
    "Actual": y_test.values,
    "Predicted": y_pred
})

pred_df.to_csv(os.path.join(output_dir, "aloa_predictions.csv"), index=False)

# ==============================
# SAVE + SHOW FUNCTION
# ==============================
def save_and_show(fig, filename):
    path = os.path.join(output_dir, filename)
    fig.savefig(path)
    plt.show()
    print(f"Saved: {filename}")

# ==============================
# GRAPH 1: ACCURACY
# ==============================
fig1 = plt.figure()
plt.bar(["ALOA_RF"], [accuracy])
plt.title("ALOA Optimized Accuracy")
plt.ylabel("Accuracy (%)")

save_and_show(fig1, "aloa_accuracy_graph.png")

# ==============================
# GRAPH 2: PREDICTION
# ==============================
fig2 = plt.figure()
plt.plot(pred_df["Actual"].values[:50], label="Actual")
plt.plot(pred_df["Predicted"].values[:50], label="Predicted")

plt.legend()
plt.title("ALOA Prediction Graph")

save_and_show(fig2, "aloa_prediction_graph.png")

# ==============================
# GRAPH 3: HEATMAP
# ==============================
corr = df.corr(numeric_only=True)

fig3 = plt.figure()
plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("ALOA Correlation Heatmap")

save_and_show(fig3, "aloa_heatmap.png")

# ==============================
# GRAPH 4: METRICS
# ==============================
fig4 = plt.figure()
plt.bar(["MSE", "R2"], [mse, r2])

plt.title("ALOA Model Metrics")

save_and_show(fig4, "aloa_metrics_graph.png")

# ==============================
# SAVE JSON SUMMARY
# ==============================
summary = {
    "best_parameters": {
        "n_estimators": int(best_solution[0]),
        "max_depth": int(best_solution[1])
    },
    "MSE": float(mse),
    "R2": float(r2),
    "Accuracy": float(accuracy)
}

with open(os.path.join(output_dir, "aloa_summary.json"), "w") as f:
    json.dump(summary, f, indent=4)

print("\n✅ ALOA RESULTS GENERATED SUCCESSFULLY!")
```
