```python
import os
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor

# ==============================
# PATHS
# ==============================
data_path = r"C:\Users\NXTWAVE\Downloads\Company Failure Risk Prediction\Non-Government-companies-struck-off-State-wise-distribution.csv"
output_dir = r"C:\Users\NXTWAVE\Downloads\Company Failure Risk Prediction"

os.makedirs(output_dir, exist_ok=True)

# ==============================
# LOAD DATA
# ==============================
df = pd.read_csv(data_path)
df = df.dropna()

print("\n✅ Dataset Loaded\n")

# ==============================
# PREPROCESSING
# ==============================
X = df.iloc[:, :-1]
y = df.iloc[:, -1]

# Encode categorical
for col in X.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col])

# Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# ==============================
# PSO IMPLEMENTATION
# ==============================
class Particle:
    def __init__(self):
        self.n_estimators = random.randint(50, 300)
        self.max_depth = random.randint(3, 20)

        self.velocity = [0, 0]
        self.best_position = [self.n_estimators, self.max_depth]
        self.best_score = -np.inf

    def get_position(self):
        return [int(self.n_estimators), int(self.max_depth)]

def fitness(position):
    n_estimators = int(position[0])
    max_depth = int(position[1])

    model = RandomForestRegressor(
        n_estimators=n_estimators,
        max_depth=max_depth,
        random_state=42
    )

    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    return r2_score(y_test, preds)

# ==============================
# PSO PARAMETERS
# ==============================
num_particles = 10
iterations = 10
w = 0.5
c1 = 1
c2 = 2

particles = [Particle() for _ in range(num_particles)]

global_best_position = None
global_best_score = -np.inf

print("\n🚀 Running PSO Optimization...\n")

# ==============================
# PSO LOOP
# ==============================
for i in range(iterations):
    for p in particles:
        current_position = p.get_position()
        score = fitness(current_position)

        # Update personal best
        if score > p.best_score:
            p.best_score = score
            p.best_position = current_position

        # Update global best
        if score > global_best_score:
            global_best_score = score
            global_best_position = current_position

    # Update velocity and position
    for p in particles:
        r1, r2 = random.random(), random.random()

        p.velocity[0] = (
            w * p.velocity[0]
            + c1 * r1 * (p.best_position[0] - p.n_estimators)
            + c2 * r2 * (global_best_position[0] - p.n_estimators)
        )

        p.velocity[1] = (
            w * p.velocity[1]
            + c1 * r1 * (p.best_position[1] - p.max_depth)
            + c2 * r2 * (global_best_position[1] - p.max_depth)
        )

        p.n_estimators += int(p.velocity[0])
        p.max_depth += int(p.velocity[1])

        # bounds
        p.n_estimators = max(10, min(300, p.n_estimators))
        p.max_depth = max(2, min(20, p.max_depth))

    print(f"Iteration {i+1} Best R2: {global_best_score:.4f}")

print("\n🏆 Best Parameters:", global_best_position)

# ==============================
# FINAL MODEL
# ==============================
best_model = RandomForestRegressor(
    n_estimators=global_best_position[0],
    max_depth=global_best_position[1],
    random_state=42
)

best_model.fit(X_train, y_train)
y_pred = best_model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
accuracy = r2 * 100

# ==============================
# SAVE RESULTS CSV
# ==============================
results_df = pd.DataFrame([{
    "Model": "PSO_RandomForest",
    "MSE": mse,
    "R2": r2,
    "Accuracy (%)": accuracy
}])

results_df.to_csv(os.path.join(output_dir, "pso_results.csv"), index=False)

# ==============================
# SAVE PREDICTIONS CSV
# ==============================
pred_df = pd.DataFrame({
    "Actual": y_test.values,
    "Predicted": y_pred
})

pred_df.to_csv(os.path.join(output_dir, "pso_predictions.csv"), index=False)

# ==============================
# SAVE + SHOW FUNCTION
# ==============================
def save_and_show(fig, filename):
    path = os.path.join(output_dir, filename)
    fig.savefig(path)
    plt.show()
    print(f"Saved: {filename}")

# ==============================
# GRAPH 1: ACCURACY
# ==============================
fig1 = plt.figure()
plt.bar(["PSO_RF"], [accuracy])
plt.title("PSO Optimized Model Accuracy")
plt.ylabel("Accuracy (%)")

save_and_show(fig1, "pso_accuracy_graph.png")

# ==============================
# GRAPH 2: PREDICTION
# ==============================
fig2 = plt.figure()
plt.plot(pred_df["Actual"].values[:50], label="Actual")
plt.plot(pred_df["Predicted"].values[:50], label="Predicted")

plt.legend()
plt.title("PSO Prediction Graph")

save_and_show(fig2, "pso_prediction_graph.png")

# ==============================
# GRAPH 3: HEATMAP
# ==============================
corr = df.corr(numeric_only=True)

fig3 = plt.figure()
plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("PSO Correlation Heatmap")

save_and_show(fig3, "pso_heatmap.png")

# ==============================
# GRAPH 4: METRICS
# ==============================
fig4 = plt.figure()
plt.bar(["MSE", "R2"], [mse, r2])

plt.title("PSO Model Metrics")

save_and_show(fig4, "pso_metrics_graph.png")

# ==============================
# SAVE JSON SUMMARY
# ==============================
summary = {
    "best_parameters": {
        "n_estimators": global_best_position[0],
        "max_depth": global_best_position[1]
    },
    "MSE": float(mse),
    "R2": float(r2),
    "Accuracy": float(accuracy)
}

with open(os.path.join(output_dir, "pso_summary.json"), "w") as f:
    json.dump(summary, f, indent=4)

print("\n✅ PSO RESULTS GENERATED SUCCESSFULLY!")
```
