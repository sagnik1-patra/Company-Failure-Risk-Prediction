```python
import os
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor

# ==============================
# PATHS
# ==============================
data_path = r"C:\Users\NXTWAVE\Downloads\Company Failure Risk Prediction\Non-Government-companies-struck-off-State-wise-distribution.csv"
output_dir = r"C:\Users\NXTWAVE\Downloads\Company Failure Risk Prediction"

os.makedirs(output_dir, exist_ok=True)

# ==============================
# LOAD DATA
# ==============================
df = pd.read_csv(data_path)
df = df.dropna()

print("\n✅ Dataset Loaded\n")

# ==============================
# PREPROCESSING
# ==============================
X = df.iloc[:, :-1]
y = df.iloc[:, -1]

# Encode categorical
for col in X.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col])

# Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# ==============================
# FITNESS FUNCTION
# ==============================
def fitness(position):
    model = RandomForestRegressor(
        n_estimators=int(position[0]),
        max_depth=int(position[1]),
        random_state=42
    )
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    return r2_score(y_test, preds)

# ==============================
# GWO PARAMETERS
# ==============================
n_wolves = 10
iterations = 10

wolves = np.random.uniform([50, 3], [300, 20], (n_wolves, 2))

print("\n🐺 Running Grey Wolf Optimization...\n")

# ==============================
# GWO LOOP
# ==============================
for t in range(iterations):
    fitness_vals = np.array([fitness(w) for w in wolves])

    sorted_idx = np.argsort(fitness_vals)[::-1]

    alpha = wolves[sorted_idx[0]]
    beta = wolves[sorted_idx[1]]
    delta = wolves[sorted_idx[2]]

    a = 2 - t * (2 / iterations)

    for i in range(n_wolves):
        for j in range(2):
            r1, r2 = random.random(), random.random()

            A1 = 2 * a * r1 - a
            C1 = 2 * r2
            D_alpha = abs(C1 * alpha[j] - wolves[i][j])
            X1 = alpha[j] - A1 * D_alpha

            r1, r2 = random.random(), random.random()
            A2 = 2 * a * r1 - a
            C2 = 2 * r2
            D_beta = abs(C2 * beta[j] - wolves[i][j])
            X2 = beta[j] - A2 * D_beta

            r1, r2 = random.random(), random.random()
            A3 = 2 * a * r1 - a
            C3 = 2 * r2
            D_delta = abs(C3 * delta[j] - wolves[i][j])
            X3 = delta[j] - A3 * D_delta

            wolves[i][j] = (X1 + X2 + X3) / 3

        # bounds
        wolves[i][0] = np.clip(wolves[i][0], 10, 300)
        wolves[i][1] = np.clip(wolves[i][1], 2, 20)

    print(f"Iteration {t+1} Best R2: {fitness(alpha):.4f}")

best_solution = alpha.astype(int)

print("\n🏆 Best Parameters:", best_solution)

# ==============================
# FINAL MODEL
# ==============================
model = RandomForestRegressor(
    n_estimators=int(best_solution[0]),
    max_depth=int(best_solution[1]),
    random_state=42
)

model.fit(X_train, y_train)
y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
accuracy = r2 * 100

# ==============================
# SAVE RESULTS CSV
# ==============================
results_df = pd.DataFrame([{
    "Model": "GWO_RandomForest",
    "MSE": mse,
    "R2": r2,
    "Accuracy (%)": accuracy
}])

results_df.to_csv(os.path.join(output_dir, "gwoa_results.csv"), index=False)

# ==============================
# SAVE PREDICTIONS CSV
# ==============================
pred_df = pd.DataFrame({
    "Actual": y_test.values,
    "Predicted": y_pred
})

pred_df.to_csv(os.path.join(output_dir, "gwoa_predictions.csv"), index=False)

# ==============================
# SAVE + SHOW FUNCTION
# ==============================
def save_and_show(fig, filename):
    path = os.path.join(output_dir, filename)
    fig.savefig(path)
    plt.show()
    print(f"Saved: {filename}")

# ==============================
# GRAPH 1: ACCURACY
# ==============================
fig1 = plt.figure()
plt.bar(["GWO_RF"], [accuracy])
plt.title("GWO Optimized Accuracy")
plt.ylabel("Accuracy (%)")

save_and_show(fig1, "gwoa_accuracy_graph.png")

# ==============================
# GRAPH 2: PREDICTION
# ==============================
fig2 = plt.figure()
plt.plot(pred_df["Actual"].values[:50], label="Actual")
plt.plot(pred_df["Predicted"].values[:50], label="Predicted")

plt.legend()
plt.title("GWO Prediction Graph")

save_and_show(fig2, "gwoa_prediction_graph.png")

# ==============================
# GRAPH 3: HEATMAP
# ==============================
corr = df.corr(numeric_only=True)

fig3 = plt.figure()
plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("GWO Correlation Heatmap")

save_and_show(fig3, "gwoa_heatmap.png")

# ==============================
# GRAPH 4: METRICS
# ==============================
fig4 = plt.figure()
plt.bar(["MSE", "R2"], [mse, r2])

plt.title("GWO Model Metrics")

save_and_show(fig4, "gwoa_metrics_graph.png")

# ==============================
# SAVE JSON
# ==============================
summary = {
    "best_parameters": {
        "n_estimators": int(best_solution[0]),
        "max_depth": int(best_solution[1])
    },
    "MSE": float(mse),
    "R2": float(r2),
    "Accuracy": float(accuracy)
}

with open(os.path.join(output_dir, "gwoa_summary.json"), "w") as f:
    json.dump(summary, f, indent=4)

print("\n✅ GWO RESULTS GENERATED SUCCESSFULLY!")
```
