```python
import os
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor

# ==============================
# PATHS
# ==============================
data_path = r"C:\Users\NXTWAVE\Downloads\Company Failure Risk Prediction\Non-Government-companies-struck-off-State-wise-distribution.csv"
output_dir = r"C:\Users\NXTWAVE\Downloads\Company Failure Risk Prediction"

os.makedirs(output_dir, exist_ok=True)

# ==============================
# LOAD DATA
# ==============================
df = pd.read_csv(data_path)
df = df.dropna()

print("\n✅ Dataset Loaded\n")

# ==============================
# PREPROCESSING
# ==============================
X = df.iloc[:, :-1]
y = df.iloc[:, -1]

# Encode categorical
for col in X.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col])

# Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# ==============================
# FITNESS FUNCTION
# ==============================
def fitness(position):
    model = RandomForestRegressor(
        n_estimators=int(position[0]),
        max_depth=int(position[1]),
        random_state=42
    )
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    return r2_score(y_test, preds)

# ==============================
# WOA PARAMETERS
# ==============================
n_whales = 10
iterations = 10

whales = np.random.uniform([50, 3], [300, 20], (n_whales, 2))

print("\n🐋 Running Whale Optimization Algorithm...\n")

# ==============================
# WOA LOOP
# ==============================
for t in range(iterations):
    fitness_vals = np.array([fitness(w) for w in whales])

    best_idx = np.argmax(fitness_vals)
    best_whale = whales[best_idx]

    a = 2 - t * (2 / iterations)

    for i in range(n_whales):
        r1, r2 = random.random(), random.random()
        A = 2 * a * r1 - a
        C = 2 * r2
        p = random.random()

        if p < 0.5:
            if abs(A) < 1:
                D = abs(C * best_whale - whales[i])
                new_pos = best_whale - A * D
            else:
                rand_whale = whales[random.randint(0, n_whales - 1)]
                D = abs(C * rand_whale - whales[i])
                new_pos = rand_whale - A * D
        else:
            # Spiral update
            b = 1
            l = random.uniform(-1, 1)
            D = abs(best_whale - whales[i])
            new_pos = D * np.exp(b * l) * np.cos(2 * np.pi * l) + best_whale

        # Bounds
        new_pos[0] = np.clip(new_pos[0], 10, 300)
        new_pos[1] = np.clip(new_pos[1], 2, 20)

        whales[i] = new_pos

    print(f"Iteration {t+1} Best R2: {fitness(best_whale):.4f}")

best_solution = best_whale.astype(int)

print("\n🏆 Best Parameters:", best_solution)

# ==============================
# FINAL MODEL
# ==============================
model = RandomForestRegressor(
    n_estimators=int(best_solution[0]),
    max_depth=int(best_solution[1]),
    random_state=42
)

model.fit(X_train, y_train)
y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
accuracy = r2 * 100

# ==============================
# SAVE RESULTS CSV
# ==============================
results_df = pd.DataFrame([{
    "Model": "WOA_RandomForest",
    "MSE": mse,
    "R2": r2,
    "Accuracy (%)": accuracy
}])

results_df.to_csv(os.path.join(output_dir, "woa_results.csv"), index=False)

# ==============================
# SAVE PREDICTIONS CSV
# ==============================
pred_df = pd.DataFrame({
    "Actual": y_test.values,
    "Predicted": y_pred
})

pred_df.to_csv(os.path.join(output_dir, "woa_predictions.csv"), index=False)

# ==============================
# SAVE + SHOW FUNCTION
# ==============================
def save_and_show(fig, filename):
    path = os.path.join(output_dir, filename)
    fig.savefig(path)
    plt.show()
    print(f"Saved: {filename}")

# ==============================
# GRAPH 1: ACCURACY
# ==============================
fig1 = plt.figure()
plt.bar(["WOA_RF"], [accuracy])
plt.title("WOA Optimized Accuracy")
plt.ylabel("Accuracy (%)")

save_and_show(fig1, "woa_accuracy_graph.png")

# ==============================
# GRAPH 2: PREDICTION
# ==============================
fig2 = plt.figure()
plt.plot(pred_df["Actual"].values[:50], label="Actual")
plt.plot(pred_df["Predicted"].values[:50], label="Predicted")

plt.legend()
plt.title("WOA Prediction Graph")

save_and_show(fig2, "woa_prediction_graph.png")

# ==============================
# GRAPH 3: HEATMAP
# ==============================
corr = df.corr(numeric_only=True)

fig3 = plt.figure()
plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("WOA Correlation Heatmap")

save_and_show(fig3, "woa_heatmap.png")

# ==============================
# GRAPH 4: METRICS
# ==============================
fig4 = plt.figure()
plt.bar(["MSE", "R2"], [mse, r2])

plt.title("WOA Model Metrics")

save_and_show(fig4, "woa_metrics_graph.png")

# ==============================
# SAVE JSON
# ==============================
summary = {
    "best_parameters": {
        "n_estimators": int(best_solution[0]),
        "max_depth": int(best_solution[1])
    },
    "MSE": float(mse),
    "R2": float(r2),
    "Accuracy": float(accuracy)
}

with open(os.path.join(output_dir, "woa_summary.json"), "w") as f:
    json.dump(summary, f, indent=4)

print("\n✅ WOA RESULTS GENERATED SUCCESSFULLY!")
```
