```python
import os
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor

# ==============================
# PATHS
# ==============================
data_path = r"C:\Users\NXTWAVE\Downloads\Company Failure Risk Prediction\Non-Government-companies-struck-off-State-wise-distribution.csv"
output_dir = r"C:\Users\NXTWAVE\Downloads\Company Failure Risk Prediction"

os.makedirs(output_dir, exist_ok=True)

# ==============================
# LOAD DATA
# ==============================
df = pd.read_csv(data_path)
df = df.dropna()

print("\n✅ Dataset Loaded\n")

# ==============================
# PREPROCESSING
# ==============================
X = df.iloc[:, :-1]
y = df.iloc[:, -1]

# Encode categorical
for col in X.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col])

# Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# ==============================
# QPSO IMPLEMENTATION
# ==============================
num_particles = 10
iterations = 10
beta = 0.75  # contraction-expansion coefficient

# Initialize particles
particles = np.random.uniform(
    low=[50, 3],
    high=[300, 20],
    size=(num_particles, 2)
)

pbest = particles.copy()
pbest_scores = np.full(num_particles, -np.inf)

def fitness(position):
    model = RandomForestRegressor(
        n_estimators=int(position[0]),
        max_depth=int(position[1]),
        random_state=42
    )
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    return r2_score(y_test, preds)

print("\n⚛️ Running QPSO Optimization...\n")

# ==============================
# QPSO LOOP
# ==============================
for iter in range(iterations):
    for i in range(num_particles):
        score = fitness(particles[i])

        if score > pbest_scores[i]:
            pbest_scores[i] = score
            pbest[i] = particles[i].copy()

    # Global best
    gbest = pbest[np.argmax(pbest_scores)]

    # Mean best
    mbest = np.mean(pbest, axis=0)

    for i in range(num_particles):
        phi = random.random()
        u = random.random()

        p = phi * pbest[i] + (1 - phi) * gbest

        particles[i] = p + beta * np.abs(mbest - particles[i]) * np.log(1/u)

        # Bounds
        particles[i][0] = np.clip(particles[i][0], 10, 300)
        particles[i][1] = np.clip(particles[i][1], 2, 20)

    print(f"Iteration {iter+1} Best R2: {np.max(pbest_scores):.4f}")

best_params = gbest.astype(int)

print("\n🏆 Best Parameters:", best_params)

# ==============================
# FINAL MODEL
# ==============================
model = RandomForestRegressor(
    n_estimators=int(best_params[0]),
    max_depth=int(best_params[1]),
    random_state=42
)

model.fit(X_train, y_train)
y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
accuracy = r2 * 100

# ==============================
# SAVE RESULTS CSV
# ==============================
results_df = pd.DataFrame([{
    "Model": "QPSO_RandomForest",
    "MSE": mse,
    "R2": r2,
    "Accuracy (%)": accuracy
}])

results_df.to_csv(os.path.join(output_dir, "qpso_results.csv"), index=False)

# ==============================
# SAVE PREDICTIONS CSV
# ==============================
pred_df = pd.DataFrame({
    "Actual": y_test.values,
    "Predicted": y_pred
})

pred_df.to_csv(os.path.join(output_dir, "qpso_predictions.csv"), index=False)

# ==============================
# SAVE + SHOW FUNCTION
# ==============================
def save_and_show(fig, filename):
    path = os.path.join(output_dir, filename)
    fig.savefig(path)
    plt.show()
    print(f"Saved: {filename}")

# ==============================
# GRAPH 1: ACCURACY
# ==============================
fig1 = plt.figure()
plt.bar(["QPSO_RF"], [accuracy])
plt.title("QPSO Optimized Model Accuracy")
plt.ylabel("Accuracy (%)")

save_and_show(fig1, "qpso_accuracy_graph.png")

# ==============================
# GRAPH 2: PREDICTION
# ==============================
fig2 = plt.figure()
plt.plot(pred_df["Actual"].values[:50], label="Actual")
plt.plot(pred_df["Predicted"].values[:50], label="Predicted")

plt.legend()
plt.title("QPSO Prediction Graph")

save_and_show(fig2, "qpso_prediction_graph.png")

# ==============================
# GRAPH 3: HEATMAP
# ==============================
corr = df.corr(numeric_only=True)

fig3 = plt.figure()
plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("QPSO Correlation Heatmap")

save_and_show(fig3, "qpso_heatmap.png")

# ==============================
# GRAPH 4: METRICS
# ==============================
fig4 = plt.figure()
plt.bar(["MSE", "R2"], [mse, r2])

plt.title("QPSO Model Metrics")

save_and_show(fig4, "qpso_metrics_graph.png")

# ==============================
# SAVE JSON SUMMARY
# ==============================
summary = {
    "best_parameters": {
        "n_estimators": int(best_params[0]),
        "max_depth": int(best_params[1])
    },
    "MSE": float(mse),
    "R2": float(r2),
    "Accuracy": float(accuracy)
}

with open(os.path.join(output_dir, "qpso_summary.json"), "w") as f:
    json.dump(summary, f, indent=4)

print("\n✅ QPSO RESULTS GENERATED SUCCESSFULLY!")
```
