```python
import os
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression

# ==============================
# PATHS
# ==============================
data_path = r"C:\Users\NXTWAVE\Downloads\Company Failure Risk Prediction\Non-Government-companies-struck-off-State-wise-distribution.csv"
output_dir = r"C:\Users\NXTWAVE\Downloads\Company Failure Risk Prediction"

os.makedirs(output_dir, exist_ok=True)

# ==============================
# LOAD DATA
# ==============================
df = pd.read_csv(data_path)
df = df.dropna()

print("Dataset Loaded")

# ==============================
# PREPROCESSING
# ==============================
X = df.iloc[:, :-1]
y = df.iloc[:, -1]

# Encode categorical
for col in X.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col])

# Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# ==============================
# MODELS
# ==============================
models = {
    "LinearRegression": LinearRegression(),
    "DecisionTree": DecisionTreeRegressor(),
    "RandomForest": RandomForestRegressor(n_estimators=200)
}

results = []

# ==============================
# TRAIN & EVALUATE
# ==============================
for name, model in models.items():
    model.fit(X_train, y_train)
    preds = model.predict(X_test)

    mse = mean_squared_error(y_test, preds)
    r2 = r2_score(y_test, preds)

    accuracy = r2 * 100  # treating R2 as accuracy %

    results.append({
        "Model": name,
        "MSE": mse,
        "R2": r2,
        "Accuracy (%)": accuracy
    })

# Convert to DataFrame
results_df = pd.DataFrame(results)

# Save results CSV
results_csv = os.path.join(output_dir, "results.csv")
results_df.to_csv(results_csv, index=False)

print("\nModel Results:")
print(results_df)

# ==============================
# BEST MODEL
# ==============================
best_model_name = results_df.sort_values(by="Accuracy (%)", ascending=False).iloc[0]["Model"]
best_model = models[best_model_name]

print("\nBest Model:", best_model_name)

# ==============================
# PREDICTIONS CSV
# ==============================
y_pred = best_model.predict(X_test)

pred_df = pd.DataFrame({
    "Actual": y_test.values,
    "Predicted": y_pred
})

pred_csv = os.path.join(output_dir, "predictions.csv")
pred_df.to_csv(pred_csv, index=False)

# ==============================
# GRAPH 1: ACCURACY GRAPH
# ==============================
plt.figure()
plt.bar(results_df["Model"], results_df["Accuracy (%)"])
plt.title("Model Accuracy Comparison")
plt.xlabel("Models")
plt.ylabel("Accuracy (%)")
plt.xticks(rotation=30)
plt.tight_layout()

plt.savefig(os.path.join(output_dir, "accuracy_graph.png"))
plt.close()

# ==============================
# GRAPH 2: COMPARISON GRAPH (MSE vs R2)
# ==============================
plt.figure()

plt.plot(results_df["Model"], results_df["MSE"], marker='o', label="MSE")
plt.plot(results_df["Model"], results_df["R2"], marker='o', label="R2")

plt.title("Model Comparison (MSE vs R2)")
plt.xlabel("Models")
plt.legend()
plt.xticks(rotation=30)
plt.tight_layout()

plt.savefig(os.path.join(output_dir, "comparison_graph.png"))
plt.close()

# ==============================
# GRAPH 3: PREDICTION GRAPH
# ==============================
plt.figure()

plt.plot(pred_df["Actual"].values[:50], label="Actual")
plt.plot(pred_df["Predicted"].values[:50], label="Predicted")

plt.title("Actual vs Predicted (First 50)")
plt.legend()
plt.tight_layout()

plt.savefig(os.path.join(output_dir, "prediction_graph.png"))
plt.close()

# ==============================
# HEATMAP (CORRELATION)
# ==============================
corr = df.corr(numeric_only=True)

plt.figure()
plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("Correlation Heatmap")
plt.tight_layout()

plt.savefig(os.path.join(output_dir, "heatmap.png"))
plt.close()

# ==============================
# EXTRA: RESULTS GRAPH (BAR)
# ==============================
plt.figure()

plt.bar(results_df["Model"], results_df["MSE"])
plt.title("MSE Comparison")
plt.xticks(rotation=30)
plt.tight_layout()

plt.savefig(os.path.join(output_dir, "mse_graph.png"))
plt.close()

# ==============================
# SAVE JSON SUMMARY
# ==============================
summary = {
    "best_model": best_model_name,
    "results": results
}

with open(os.path.join(output_dir, "summary.json"), "w") as f:
    json.dump(summary, f, indent=4)

print("\n✅ All outputs generated successfully!")
```
