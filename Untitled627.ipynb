```python
import os
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor

# ==============================
# PATHS
# ==============================
data_path = r"C:\Users\NXTWAVE\Downloads\Company Failure Risk Prediction\Non-Government-companies-struck-off-State-wise-distribution.csv"
output_dir = r"C:\Users\NXTWAVE\Downloads\Company Failure Risk Prediction"

os.makedirs(output_dir, exist_ok=True)

# ==============================
# LOAD DATA
# ==============================
df = pd.read_csv(data_path)
df = df.dropna()

print("\n✅ Dataset Loaded\n")

# ==============================
# PREPROCESSING
# ==============================
X = df.iloc[:, :-1]
y = df.iloc[:, -1]

for col in X.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col])

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# ==============================
# FITNESS FUNCTION
# ==============================
def fitness(solution):
    solution = np.array(solution)
    model = RandomForestRegressor(
        n_estimators=int(solution[0]),
        max_depth=int(solution[1]),
        random_state=42
    )
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    return r2_score(y_test, preds)

# ==============================
# BAT ALGORITHM (BA)
# ==============================
print("\n🦇 Bat Algorithm Phase...\n")

def create_bat():
    return np.array([
        random.randint(50, 300),
        random.randint(3, 20)
    ], dtype=float)

bats = [create_bat() for _ in range(10)]
velocities = [np.zeros(2) for _ in range(10)]

best_bat = bats[0]
best_score = fitness(best_bat)

for iter in range(5):
    for i in range(len(bats)):
        freq = random.random()
        velocities[i] += (bats[i] - best_bat) * freq
        bats[i] += velocities[i]

        bats[i][0] = np.clip(bats[i][0], 10, 300)
        bats[i][1] = np.clip(bats[i][1], 2, 20)

        score = fitness(bats[i])
        if score > best_score:
            best_score = score
            best_bat = bats[i].copy()

print("BA Best:", best_bat)

# ==============================
# ALOA PHASE (refinement)
# ==============================
print("\n🐜 ALOA Phase...\n")

def random_walk(center):
    step = np.random.uniform(-5, 5, size=2)
    return np.clip(center + step, [10, 2], [300, 20])

ants = [random_walk(best_bat) for _ in range(10)]

global_best = np.array(best_bat, dtype=float)
global_best_score = fitness(global_best)

for iter in range(10):
    for i in range(len(ants)):
        ants[i] = random_walk(global_best)
        score = fitness(ants[i])

        if score > global_best_score:
            global_best_score = score
            global_best = ants[i].copy()

    print(f"Iteration {iter+1} Best R2: {global_best_score:.4f}")

best_solution = global_best.astype(int)

print("\n🏆 Hybrid BA + ALOA Best Parameters:", best_solution)

# ==============================
# FINAL MODEL
# ==============================
model = RandomForestRegressor(
    n_estimators=int(best_solution[0]),
    max_depth=int(best_solution[1]),
    random_state=42
)

model.fit(X_train, y_train)
y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
accuracy = r2 * 100

# ==============================
# SAVE RESULTS
# ==============================
results_df = pd.DataFrame([{
    "Model": "BA+ALOA Hybrid",
    "MSE": mse,
    "R2": r2,
    "Accuracy (%)": accuracy
}])

results_df.to_csv(os.path.join(output_dir, "bloa_results.csv"), index=False)

pred_df = pd.DataFrame({
    "Actual": y_test.values,
    "Predicted": y_pred
})

pred_df.to_csv(os.path.join(output_dir, "bloa_predictions.csv"), index=False)

# ==============================
# SAVE + SHOW
# ==============================
def save_and_show(fig, name):
    path = os.path.join(output_dir, name)
    fig.savefig(path)
    plt.show()
    print(f"Saved: {name}")

# Accuracy Graph
fig1 = plt.figure()
plt.bar(["BA+ALOA"], [accuracy])
plt.title("Hybrid BA + ALOA Accuracy")
save_and_show(fig1, "bloa_accuracy_graph.png")

# Prediction Graph
fig2 = plt.figure()
plt.plot(pred_df["Actual"].values[:50], label="Actual")
plt.plot(pred_df["Predicted"].values[:50], label="Predicted")
plt.legend()
plt.title("Prediction Graph")
save_and_show(fig2, "bloa_prediction_graph.png")

# Heatmap
corr = df.corr(numeric_only=True)
fig3 = plt.figure()
plt.imshow(corr)
plt.colorbar()
plt.title("Heatmap")
save_and_show(fig3, "bloa_heatmap.png")

# Metrics
fig4 = plt.figure()
plt.bar(["MSE", "R2"], [mse, r2])
plt.title("Metrics")
save_and_show(fig4, "bloa_metrics_graph.png")

# JSON
summary = {
    "best_parameters": {
        "n_estimators": int(best_solution[0]),
        "max_depth": int(best_solution[1])
    },
    "MSE": float(mse),
    "R2": float(r2),
    "Accuracy": float(accuracy)
}

with open(os.path.join(output_dir, "bloa_summary.json"), "w") as f:
    json.dump(summary, f, indent=4)

print("\n✅ HYBRID BA + ALOA COMPLETED SUCCESSFULLY!")
```
